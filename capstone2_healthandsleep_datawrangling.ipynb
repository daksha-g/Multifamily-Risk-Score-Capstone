## Imports<a id= 'Imports'></a>

```python
import pandas as pd
import numpy as np
```

## Load Data<a id='Load_Data'></a>

```python
df = pd.read_csv("health_sleep_relation_2024.csv")
print(df.head())
```

```python
## Data Definition<a id='Data_Definition'></a>


#I look at the data types to make sure they make sense
print("\nData Types:\n", df.dtypes)

#summary statistics (mean, median, etc.) for numeric columns
print("\nSummary Statistics:\n", df.describe())

#counting  of unique values in categorical columns
for col in ["Gender", "Occupation", "BMI Category", "Sleep_Disorder"]:
    print(f"\nUnique values in {col}:")
    print(df[col].value_counts())
```

```python
## Data Organization <a id='Data_Organization'></a>


#renaming columns to make them easier to understand, for example I changed "PhysicalActivityLevel" to "Activity_Level")
df = df.rename(columns={
    "Sleep Duration": "Sleep_Duration",
    "Quality of Sleep": "Sleep_Quality",
    "Physical Activity Level": "Activity_Level",
    "Stress Level": "Stress_Level",
    "Heart Rate": "Heart_Rate",
    "Daily Steps": "Daily_Steps",
    "Sleep Disorder": "Sleep_Disorder"
})

#this line prints new column names
print("Updated Column Names:\n", df.columns)
```

```python
## Data Cleaning<a id='Data_Cleaning'></a>

#checking for missing values
print("\nMissing values in each column:\n", df.isnull().sum())

#drops duplicate rows 
df = df.drop_duplicates()
print("\nShape after dropping duplicates:", df.shape)

#fills missing categorical values (like Sleep_Disorder) with "None"
df["Sleep_Disorder"] = df["Sleep_Disorder"].fillna("None")

#handles missing numeric values by filling with column mean
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())

#confirms no missing values left
print("\nMissing values after cleaning:\n", df.isnull().sum())

#this is an example - groups average sleep hours by occupation same with next line
occupation_sleep = df.groupby("Occupation")["Sleep_Duration"].mean()
print("\nAverage Sleep Hours by Occupation:\n", occupation_sleep)
sleep_disorder_counts = df["Sleep_Disorder"].value_counts()
print("\nSleep Disorder Counts:\n", sleep_disorder_counts)

# stores cleaned data for later use
df.to_csv("Health_and_Sleep_Cleaned.csv", index=False)
print("\nCleaned dataset saved as Health_and_Sleep_Cleaned.csv")
```
